```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('bengaluru_house_prices.csv')
```

```python
data.head(10)
```

```python
data['area_type'].unique()
```

```python
data['availability'].count()
```

```python
data.shape
```

```python
data.drop(columns=['area_type', 'availability', 'society', ], inplace=True)
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data['location'] = data['location'].apply(lambda x: x.strip())
data['location'].value_counts()
```

```python
data.location.value_counts()
```

```python
location_stats = data.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_less_than_10 = location_stats[location_stats <= 10]
location_less_than_10
```

```python
data['location'] = data['location'].replace(location_less_than_10.index, 'other')
data['location'].value_counts()
```

```python
data
```

```python
data['total_sqft'].unique()
```

```python
data['size'].unique()
```

```python
data['bedrooms'] = data['size'].str.split(' ').str[0].astype(int)
```

```python
data
```

```python
data['total_sqft'].unique()
```

```python
def clean(Sqft):
    tokrns = Sqft.split('-')
    if len(tokrns) == 2:
        return (float(tokrns[0]) + float(tokrns[1])) / 2
    else:
        try:
            return float(Sqft)
        except:
            return None

            
```

```python
data['total_sqft'] = data['total_sqft'].apply(clean)
data['total_sqft'].unique()
```

```python
data.describe()
```

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data ['sqft_per_bed'] = data['total_sqft'] / data['bedrooms']
```

```python
data.sqft_per_bed.describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
data2
```

```python
data2['price_per_sqft'] = round(data['price'] * 100000 / data2['total_sqft'],2)
data2
```

```python
data2['price_per_sqft'].describe()
```

```python
data3 = data2[data2['price_per_sqft'] >=2000]
data3
```

```python
data3.drop(columns=['size','sqft_per_bed' ,'price_per_sqft'], axis=1 , inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
col_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']),remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(col_trans, scaler, lr)
```

```python
data_input = data3.drop(columns=['price'])
data_output = data3['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
input = pd.DataFrame([['Electronic City Phase II', 1500.0, 2.0, 3.0, 3.0]], 
                     columns=['location', 'total_sqft', 'bath', 'balcony', 'bedrooms'])
```

```python
model.predict(input)
```

```python
import pickle as pk
pk.dump(model, open('House_prediction_model.pkl', 'wb'))
```
